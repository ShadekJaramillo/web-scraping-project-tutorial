```python
import app
import matplotlib.pyplot as plt
url = 'https://listado.mercadolibre.com.co/arte-papeleria-merceria/papeleria/papeleria_Desde_0_NoIndex_True'
items_html = app.get_items_mercadolibre(url)
df = app.create_dataframe(items_html)
df.head(10).style.hide(axis='index')
```

```python
fig, ax = plt.subplots(3, 1)

ax[0].hist(df.price, bins = 20)
ax[0].set_title('price distribution for stationery items')
ax[0].set_xlabel('price (COP)')

ax[1].scatter(df.price, df.discount_percentage)
ax[1].set_title('relation between price and discount')
ax[1].set_xlabel('price (COP)')
ax[1].set_ylabel('discount %')

fig.

plt.show()
```
